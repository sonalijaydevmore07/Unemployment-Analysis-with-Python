```python
import pandas as pd
#reading the data
data = pd.read_csv("spam.csv") #,encoding="ISO-8859-1")
data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
data_spam = data[data['v1']=='spam'][['v2']]
data_spam1 = data[data['v1']=='ham'][['v2']]
data_spam
```

```python
text = " ".join(cat.split()[1] for cat in data_spam.v2)
text
```

```python
text1 = " ".join(cat.split()[0] for cat in data_spam1.v2)
text1 
```

```python
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text)
```

```python
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text1)
```

```python
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
